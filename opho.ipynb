6. Polarizers (Correct)

```python
def efficiency6(a,b,c):
    return (np.cos(a)*np.cos(b)*np.cos(c))**2

x=np.linspace(0,np.pi/2,100)
y=np.pi/2 -x
X, Y = np.meshgrid(x, y)
Z=np.pi/2 -X-Y
d = efficiency6(X,Y,Z)
Z[Z < 0] = np.nan
color_dimension = d

minn, maxx = color_dimension.min(), color_dimension.max()
norm = Normalize(minn, maxx)
m = plt.cm.ScalarMappable(norm=norm, cmap='jet')
m.set_array([])
fcolors = m.to_rgba(color_dimension)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=fcolors,linewidth=0)#, antialiased=False)
ax.set_xlim(0, np.pi/2)
ax.set_ylim(0, np.pi/2)
ax.set_zlim(0, np.pi/2)
ax.view_init(elev=45, azim=45)

plt.show()
```

23. Resistance box (Correct)

```python
a=1/3
b=4/3
c=2/3
d=1/3
e=1

s,p1,p2=a,b,c

n=10

s = s+ p1*p2/(p1+p2+d)
p1_temp = b*d/(p1+p2+d)
p2 = c*d/(p1+p2+d)
p1 = p1_temp

for i in range(n):
    p1 = p1+e
    p2 = p2+c
    s = s+ p1*p2/(p1+p2+d)
    p1_temp = p1*d/(p1+p2+d)
    p2 = p2*d/(p1+p2+d)
    p1 = p1_temp
p1 = p1+e
p2 = p2+b
s = s+ p1*p2/(p1+p2)
print(s/n)
```

25. Dying Light (Incorrect)

```python
n=10000
phi=np.pi/(2*n)
full_theta = 1#np.pi
phi=full_theta/n
f=100
lam=f/n
x2=0
theta=[]
x=[]
x1_0=lam/2
#print(x1)
theta1=0
x1=x1_0
for i in range(n):#int(n*np.pi/2)):
    theta1 = abs(theta1 - x1/f)
    x2 = x1/(np.sin(phi)*np.tan(theta1)+np.cos(phi))
    theta2 = -(phi - theta1)
    theta.append(theta1)
    x.append(x1)
    x1=x2
    theta1=theta2
print(x1/(x1_0),sum(theta),full_theta,180*sum(theta)/np.pi,sum(x)/(lam/2))
```
